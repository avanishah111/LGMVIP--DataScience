## Image to pencil sketch using python:

Here are the steps to perform the image-to-pencil-sketch conversion:

1. Read the image.
2. Convert the image to grayscale.
3. Invert the grayscale image.
4. Apply Gaussian blur to the inverted image.
5. Create the pencil sketch by dividing the grayscale image by the inverted blurred image.
6. Display the original image and the pencil sketch side by side.
7. Optionally, save the pencil sketch to a file.

These steps outline the process of converting an image to a pencil sketch using the provided code.

### importing the required libraries:

```python
import cv2
import numpy as np
from IPython.display import Image, display
```

### Specify the path to your image file & display the original image

```python
image_path = "E:\imp\lion.png"

# Read the image
image = cv2.imread(image_path)

# Check if the image was successfully loaded
if image is not None:
    # Image was loaded successfully
    # Display the image inline in the notebook
    display(Image(filename=image_path))
else:
    # Failed to load image
    print(f"Error: Failed to load image '{image_path}'")
```

```python
# Read the image in RGB format
image = cv2.imread(image_path)
image1 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display(Image(data=cv2.imencode('.png', image1)[1].tobytes()))
```

```python
# Convert the image to grayscale
grayscale_image = cv2.cvtColor(image1, cv2.COLOR_RGB2GRAY)
display(Image(data=cv2.imencode('.png', grayscale_image)[1].tobytes()))
```

```python
# Invert the grayscale image
inverted_image = cv2.bitwise_not(grayscale_image)
display(Image(data=cv2.imencode('.png', inverted_image)[1].tobytes()))
```

```python

# Apply Gaussian blur to the inverted image
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
display(Image(data=cv2.imencode('.png', blurred_image)[1].tobytes()))
```

### Sketch of image

```python
# Create the pencil sketch by dividing the grayscale image by the inverted blurred image
pencil_sketch = cv2.divide(grayscale_image, blurred_image, scale=256.0)
display(Image(data=cv2.imencode('.png', pencil_sketch)[1].tobytes()))
```

```python
# Display the original image and pencil sketch
display(Image(filename=image_path))

display(Image(data=cv2.imencode('.png', pencil_sketch)[1].tobytes()))
```

### Save image in your pc:

```python
# Specify the path and filename to save the sketch image

sketch_image_path = "E:\imp\lion_sketch.png"

# Save the sketch image

cv2.imwrite(sketch_image_path, pencil_sketch)

# The cv2.imwrite() function returns a boolean value indicating whether the image was successfully written to the file or not. 
# If the function call returns True, it means the image was saved successfully.
#If it returns False, there may have been an issue with the path or filename specified.
```

### ANOTHER METHOD USING PYTHON

```python
import cv2
import numpy as np
from IPython.display import Image, display

def pencil_sketch(image_path):
    # Read the image in RGB format
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Convert the image to grayscale
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    
    # Invert the grayscale image
    inverted_image = cv2.bitwise_not(grayscale_image)
    
    # Apply Gaussian blur to the inverted image
    blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
    
    # Create the pencil sketch by dividing the grayscale image by the inverted blurred image
    pencil_sketch = cv2.divide(grayscale_image, blurred_image, scale=256.0)
    
    # Display the original image and the pencil sketch
    display(Image(filename=image_path))
    display(Image(data=cv2.imencode('.png', pencil_sketch)[1].tobytes()))

# Specify the path to your image file
image_path ="E:\imp\lion.png"


# Generate the pencil sketch
pencil_sketch(image_path)
```
